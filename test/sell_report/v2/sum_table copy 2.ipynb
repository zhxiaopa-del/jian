# 读取明细表数据

读取明细表，获取回款和合同两个表的数据

```python
from pathlib import Path
from openpyxl import load_workbook

PROJECT_ROOT = Path.cwd()
DETAIL_TABLE_PATH = PROJECT_ROOT / "data" / "明细表.xlsx"
```

```python
def read_detail_table():
    """读取明细表，返回两个表的数据"""
    if not DETAIL_TABLE_PATH.exists():
        return [], []
    
    wb = load_workbook(DETAIL_TABLE_PATH, read_only=True, data_only=True)
    
    # 存储两个表的数据
    payment_data = []  # 回款数据
    contract_data = []  # 合同数据
    
    for sheet_idx, sheet_name in enumerate(wb.sheetnames, 1):
        ws = wb[sheet_name]
        
        # 读取所有数据行
        sheet_data = []
        for row in ws.iter_rows(min_row=2, values_only=True):
            if any(row):  # 只添加非空行
                sheet_data.append(list(row))
        
        # 根据工作表名称或索引判断是回款还是合同
        if sheet_idx == 1 or "回款" in sheet_name:
            payment_data = sheet_data
        elif sheet_idx == 2 or "合同" in sheet_name:
            contract_data = sheet_data
    
    wb.close()
    
    return payment_data, contract_data
```

```python
# 读取数据
payment_data, contract_data = read_detail_table()

print(f"回款数据条数: {len(payment_data)}")
print(f"合同数据条数: {len(contract_data)}")
```
